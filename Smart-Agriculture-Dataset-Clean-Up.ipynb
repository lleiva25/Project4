```python
#Paths for files
agriculture_path = '/kaggle/input/smart-agricultural-production-optimizing-engine/Crop_recommendation.csv'


#Read csv files and create dataframes
agriculture_df = pd.DataFrame(pd.read_csv(agriculture_path,skiprows=0))

#See 5 rows from df
agriculture_df.head()
```

```python
print('Unique Counts of Nitrogen')
print('='*50)
print(agriculture_df['N'].value_counts())


print('Unique Counts of Phosphate')
print('='*50)
print(agriculture_df['P'].value_counts())

print('='*50)
print('Unique types of Potassium')
print(agriculture_df['K'].value_counts())

print('Unique Counts of Temperature')
print('='*50)
print(agriculture_df['temperature'].value_counts())


print('Unique Counts of Humidity')
print('='*50)
print(agriculture_df['humidity'].value_counts())

print('='*50)
print('Unique types of pH')
print(agriculture_df['ph'].value_counts())

print('='*50)
print('Unique types of rainfall')
print(agriculture_df['rainfall'].value_counts())

print('='*50)
print('Unique types of Labels')
print(agriculture_df['label'].value_counts())
```
