```python
#Create a dataframe from the read csv
df_2017 = pd.DataFrame(pd.read_csv(path_2017, skiprows=1))
df_2018 = pd.DataFrame(pd.read_csv(path_2018, skiprows=1))
df_2019 = pd.DataFrame(pd.read_csv(path_2019, skiprows=1))

df_2017.head()
```
