```python
#Paths for files
commodities_path = '/kaggle/input/global-food-prices/wfp_commodities_global.csv'
countries_path = '/kaggle/input/global-food-prices/wfp_countries_global.csv'
markets_path = '/kaggle/input/global-food-prices/wfp_markets_global.csv'

#Read csv files and create dataframes
comm_df = pd.DataFrame(pd.read_csv(commodities_path,skiprows=1))
countries_df = pd.DataFrame(pd.read_csv(countries_path,skiprows=1))
markets_df = pd.DataFrame(pd.read_csv(markets_path,skiprows=1))

#See 5 rows from df
comm_df.head()
```

```python
print('Unique Counts of Item Code for Commodities')
print('='*50)
print(comm_df['#item+code'].value_counts())


print('Unique Counts of Item Type for Commodities')
print('='*50)
print(comm_df['#item+type'].value_counts())

print('='*50)
print('Unique types of Item Name for Commodities')
print(comm_df['#item+name'].value_counts())
```

```python
print('Unique Counts of Country Code for CountriesDF')
print('='*50)
print(countries_df['#country+code'].value_counts())


print('Unique Counts of Country Url for CountriesDF')
print('='*50)
print(countries_df['#country+url'].value_counts())

print('='*50)
print('Unique types Date Start for CountriesDF')
print(countries_df['#date+start'].value_counts())

print('='*50)
print('Unique types Date End for CountriesDF')
print(countries_df['#date+end'].value_counts())
```
